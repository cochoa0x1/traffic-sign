```python
import cv2 as cv2
from os import listdir
```

```python
images=[]
for f in listdir('traffic_images'):
    if f.split('.')[-1] not in ['jpeg','png','jpg']:
        continue
    print(f)
    
    im = cv2.imread('traffic_images/%s'%f)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    im = cv2.resize(im,(32,32))
    images.append(im)
```

```python
%pylab inline
```

```python
import pickle
pickle.dump( images, open( "test_images.p", "wb" ) )
```
